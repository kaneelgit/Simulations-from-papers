```python
import numpy as np
import matplotlib.pyplot as plt
from random import randrange, uniform
```

```python
#scatter points 

n_trials = 10
x = 0
xhatv = []
xhata = []

for t in range(n_trials):
    
    #noise
    epsilon_v = uniform(-1,1)*2
    epsilon_a = uniform(-1,1)*5
    
    
    #internal estimate
    x_hatv = x + epsilon_v
    x_hata = x + epsilon_a
    
    #append
    xhatv.append(x_hatv)
    xhata.append(x_hata)

plt.xkcd()
plt.figure(figsize=(8, 8))
plt.scatter(0, 0, label = 'Target', s = 100)
plt.scatter(xhatv, np.zeros(len(xhatv)), label = 'Visual estimate in each trial')
plt.scatter(xhata, np.zeros(len(xhata)), label = 'Auditory estimate in each trial')
plt.legend()
plt.show()
    
    
    
    
    
    
```
