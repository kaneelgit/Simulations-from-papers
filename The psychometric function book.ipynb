```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

**ADAPTIVE METHODS**

Transformed Up/Down Method

```python
#stimuli 
s = np.arange(0, 1, 0.01)
mu = 0.5
sigma = 0.1

#simulate an observer 
P = 1/2 + (1/2) * stats.norm.cdf(s, mu, sigma)

#plot the observer
plt.xkcd()
plt.plot(s, P)
plt.title('Observer')
plt.xlabel('Stimuli')
plt.ylabel('Probability')
plt.show()
```

```python
#number of trials
nTrials = 50

s0 = 1
step_size = 0.1


#stimulus
stim = []
val = []

for n in range(nTrials):
    
    #response probability
    rp = stats.norm.cdf(s0, mu, sigma)
    
    #response of the first trial
    response = np.random.choice([0, 1], p = [1-rp, rp])
    
    
    if response == 1 and rp >= 0.75:
        s0 -= step_size 
        v = 1
        
    elif response == 0 and rp > 0.5:
        s0 += step_size 
        v = 0
        
    stim.append(s0)
    val.append(v)
```

```python
#plot the response
plt.figure(figsize = (15, 6))
plt.plot(range(nTrials), stim)
plt.xlabel('Trial')
plt.ylabel('stimulus')
for i in range(nTrials):
    
    if val[i] == 1:
        plt.scatter(i, stim[i], color = 'b', s = 100)
    
    elif val[i] == 0:
        plt.scatter(i, stim[i], color = 'b', facecolor = 'none', s = 100)
    
```
