```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
#testing
def my_gaussian(sigma, mu,x):
    
    gaussian = 1/(np.sqrt(sigma*2*np.pi))*np.exp(-1/2*(x - mu)**2/sigma)
    gaussian = gaussian/np.sum(gaussian)
    
    return gaussian

def my_psycho_curve(gaussian):
    
    curve = np.zeros(len(gaussian))
    curve[0] = gaussian[0]
    
    for i in range(len(gaussian)-1):

        curve[i+1] = curve[i] + gaussian[i+1]       
    
    return curve


x = np.arange(-20,20,0.1)

#sigma
sigma_4 = 4
sigma_32 = 32
sigma_64 = 64

#mu
mu = 0

#create the gaussian for visual condition with variance 4
gaussian_v4 = my_gaussian(sigma_4, mu, x)
gaussian_v32  = my_gaussian(sigma_32, mu, x)
gaussian_v64 = my_gaussian(sigma_64, mu , x)


#plot the psychometric curves
psycho_v4 = my_psycho_curve(gaussian_v4)
psycho_v32 = my_psycho_curve(gaussian_v32)
psycho_v64 = my_psycho_curve(gaussian_v64)


def interact(mu_w = 0, sigma_w = 10):
    gaussian = my_gaussian(sigma_w, mu_w, x)
    psycho_curve = my_psycho_curve(gaussian)
    plt.plot(x, psycho_v4)
    plt.plot(x, psycho_v32)
    plt.plot(x, psycho_v64)
    plt.plot(x, psycho_curve)
    plt.show()
    
_ = widgets.interact(interact, mu_w = (-10,10, 0.5), sigma_w = (1, 200, 0.5))
    
```

```python
plt.style.use('grayscale')

plt.figure(figsize = (8, 6))
plt.plot(x, psycho_v4, linewidth = 3)
plt.xlabel('$Y_R$', fontsize = 12)
plt.ylabel('Proportion Correct R is more further away body', fontsize = 12)
```

Unimodal - 

In this case, the subject is presented with two stimulus consequently. In the <u>vision condition</u>, the subject is shown the standard stimulus located at the center and the comparison stimulus that varies during the experiment. The visual condtion has 3 different gaussian blobs with different amounts of noise added. 

In the <u>auditory condition</u>, again a standard and a comparision stimulus is presented to the subject. The standard stimulus is always at the center and the comparison stimulus changes during trials. 

```python
#probe location
x = np.arange(-20,20,0.1)
#sigma visual
sig_4, sig_32, sig_64 = 2, 16, 32
#mean
mu = 0
#sigma auditory
sig_au = 20

#create gaussian
visual_4 = my_gaussian(sig_4, mu, x)
visual_32 = my_gaussian(sig_32, mu, x)
visual_64 = my_gaussian(sig_64, mu, x)
auditory = my_gaussian(sig_au, mu, x)

#create psychometric curve
psycho_v4 = my_psycho_curve(visual_4)
psycho_v32 = my_psycho_curve(visual_32)
psycho_v64 = my_psycho_curve(visual_64)
psycho_a = my_psycho_curve(auditory)


plt.figure(figsize = (8,4))
plt.plot(x, np.flip(psycho_v4), label = 'v4', color = 'k')
plt.plot(x, np.flip(psycho_v32), label = 'v32', color = 'r')
plt.plot(x, np.flip(psycho_v64), label = 'v64', color = 'b')
plt.plot(x, np.flip(psycho_a), label = 'au', color = 'g')
plt.vlines(x[190], 0, np.flip(psycho_v4)[190], linestyle="dashed")
plt.hlines(np.flip(psycho_v4)[190], -20, x[190], linestyle="dashed")
plt.hlines(np.flip(psycho_v4)[197], -20, x[197], linestyle = 'dashed', color = 'red')
plt.legend()
plt.title('Figure 1-A')
plt.xlabel('Displacement of Probe')
plt.ylabel('Prop of probe seen towards left')
plt.show()
```

*Bimodal -* 

In the Bimodal condition visual and auditory stimulus are shown simultaneously. The subject is shown two presentations in each trial. In the first presentation there is a conflict between the two stimuli. The visual stimulus is shifted to the right and the auditory stimulus is shifted to the left by the same amount($\Delta$). The second presentation is a non-conflict presentation. In this one, both visual and auditory stimuli is shifted either to the left or to the right by the same amount. 

*Equations used in the bimodal condition -*

Optimal combination of the independent auditory and visual estimates $S_V$ and $S_V$ are given by the following equation. 

$$ S_{VA} = w_vS_v + w_AS_A$$

Where $S_{VA}$ is the bimodal estimate and $ w_A = \frac{\sigma^2_V}{\sigma^2_A + \sigma^2_V}$

According to the equation each modality is weighted by it's precision. ($\frac{1}{\sigma^2}$)



The variancee of the combined estimate will always be less than either individual estimate. The variance of the optimal combination is measured by the following equation. 

$$\sigma^2_{VA} = \frac{\sigma^2_V\sigma^2_A}{\sigma^2_A + \sigma^2_V}$$

**Figure 1-B**

```python
#calculate the optimal combined location
def opt_loc(sigma_visual, sigma_auditory, sv, sa):
    term1 = (sigma_auditory**2/(sigma_auditory**2 + sigma_visual**2))*sv
    term2 = (sigma_visual**2/(sigma_auditory**2 + sigma_visual**2))*sa
    return term1 + term2

#calculate the optimal variance
def opt_var(sigma_visual, sigma_auditory):
    numerator = sigma_visual**2*sigma_auditory**2
    den = sigma_visual**2 + sigma_auditory**2
    return numerator/den

#sigma of visual and auditory
sigma_visual = 2
sigma_auditory = 20

#first look at when delta = 5
S_V, S_A = 5, -5

#calculate mean location and sigma
S1_va = opt_loc(sigma_visual, sigma_auditory, S_V, S_A)
sigma1_va = opt_var(sigma_visual, sigma_auditory)

gaussian_delta5= my_gaussian(sigma1_va, S1_va, x)
delta5_psycho = my_psycho_curve(gaussian_delta5)

#look at when delta = 2.5
S2_v, S2_a = 2.5, -2.5

S2_va = opt_loc(sigma_visual, sigma_auditory, S2_v, S2_a)
sigma2_va = opt_var(sigma_visual, sigma_auditory)

gaussian_delta2 = my_gaussian(sigma2_va, S2_va, x)
delta2_psycho = my_psycho_curve(gaussian_delta2)

#look when delta = 0
S3_v, S3_a = 0, 0

S3_va = opt_loc(sigma_visual, sigma_auditory, S3_v, S3_a)
sigma3_va = opt_var(sigma_visual, sigma_auditory)

gaussian_delta0 = my_gaussian(sigma3_va, S3_va, x)
delta0_psycho = my_psycho_curve(gaussian_delta0)

#look when delta = -2.5
S4_v, S4_a = -2.5, 2.5

S4_va = opt_loc(sigma_visual, sigma_auditory, S4_v, S4_a)
sigma4_va = opt_var(sigma_visual, sigma_auditory)

gaussian_delta_2 = my_gaussian(sigma4_va, S4_va, x)
delta_2_psycho = my_psycho_curve(gaussian_delta_2)

#look when delta = -5
S5_v, S5_a = -5, 5

S5_va = opt_loc(sigma_visual, sigma_auditory, S5_v, S5_a)
sigma5_va = opt_var(sigma_visual, sigma_auditory)

gaussian_delta_5 = my_gaussian(sigma5_va, S5_va, x)
delta_5_psycho = my_psycho_curve(gaussian_delta_5)

#append all the PSE and conflict
pse4 = [(5, S1_va), (2.5, S2_va), (0, S3_va), (-2.5, S4_va), (-5, S5_va)]

#localization error
loc_error4 = (4, np.sqrt(sigma5_va))


plt.figure(figsize = (8,4))
plt.title('Figure 1-B - Bimodal $4\degree$')
plt.plot(x, np.flip(delta5_psycho), color = 'm', label = '5')
plt.plot(x, np.flip(delta2_psycho), color = 'b', label = '2.5')
plt.plot(x, np.flip(delta0_psycho), color = 'g', label = '0')
plt.plot(x, np.flip(delta_2_psycho), color = 'orange', label = '-2.5')
plt.plot(x, np.flip(delta_5_psycho), color = 'k', label = '-5')
plt.legend()
plt.xlabel('Displacement of probe')
plt.ylabel('Prop of probe seen towards left')
plt.show()
```

The x axis of this figure shows the displacement of the non conflict presentation (Probe) throughout the bimodal condition. The y axis shows the proportion a subject reports that the second stimulus was more towards the left. 

The magenta curve shows the condition where the conflicted presentation is shifting the visual stimulus $5^{\circ}$ to the right and the auditory stimulus $5^{\circ}$ to the left   ($\Delta5$). The PSE (point where the proportion is 50\%) of the magenta curve is when the displacement of the probe is at about $-5^{\circ}$. This shows that the subjects are biased towards the visual stimulus until it is obvious that the non-conflict stimulus (Probe) is actually towards the left. This shows the "ventriloquist effect"
